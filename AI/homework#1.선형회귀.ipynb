```python
x = [165, 181, 170, 173, 158, 162, 183, 174, 183, 154, 165, 163, 151, 177, 191, 181, 170, 165, 172, 156]
y = [56, 66, 60, 61, 55, 67, 92, 68, 80, 42, 71, 63, 44, 64, 95, 73, 55, 56, 83, 60]

# 균등 분포를 가진 무작위 값
W = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b = tf.Variable(tf.random_uniform([1], -1.0, 1.0))

X = tf.placeholder(tf.float32, name="Height")
Y = tf.placeholder(tf.float32, name="Weight")
```

```python
hypothesis = W * X + b

cost = tf.reduce_mean(tf.square(hypothesis - Y))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.00001)
train_op = optimizer.minimize(cost)
```

```python
with tf.Session() as sess:
  sess.run(tf.global_variables_initializer())

  for step in range(300):
    _, cost_val = sess.run([train_op, cost], feed_dict={X: x, Y: y})

    print(step, cost_val, sess.run(W), sess.run(b))

  print("\n===Test===")
  print("X: 179, Y:", sess.run(hypothesis, feed_dict={X: 179}))
  print("X: 166, Y:", sess.run(hypothesis, feed_dict={X: 166}))
  print("X: 195, Y:", sess.run(hypothesis, feed_dict={X: 195}))
```
